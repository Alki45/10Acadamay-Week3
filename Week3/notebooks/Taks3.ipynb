```python
import pandas as pd
from scipy.stats import chi2_contingency, ttest_ind
```

```python

# Load your dataset
data = pd.read_csv('A:/10x/10Acadamay-Week3/Data_Ver/Insurance_Data.csv')

```

```python
# Get unique value counts for all columns
unique_value_counts = {col: data[col].nunique() for col in data.columns}

# Print the unique value counts for each column
for col, count in unique_value_counts.items():
    if(count>1 and count<=3):
        print(f"Column '{col}' : {count} ")
```

```python
# Select relevant columns for analysis
columns_of_interest = ['AlarmImmobiliser', 'Gender'] 

```

```python
# Data Segmentation
group_a = data[data['AlarmImmobiliser'] == 'Yes']  
group_b = data[data['AlarmImmobiliser'] == 'No']  

```

```python
# Statistical Testing
for column in columns_of_interest:
    if data[column].dtype == 'object':
        # Categorical data - Chi-squared test
        contingency_table = pd.crosstab(data[column], data['Gender'])
        chi2, p, _, _ = chi2_contingency(contingency_table)
        print(f"Chi-squared test p-value for '{column}': {p}")
        
        # Analyze p-value to reject/nullify the null hypothesis
        
    else:
        # Numerical data - T-test
        t_stat, p = ttest_ind(group_a[column], group_b[column])
        print(f"T-test p-value for '{column}': {p}")
```

```python
column_pairs = {
    'Gender': 'AlarmImmobiliser',
    'TrackingDevice': 'NewVehicle'
}

for column, kpi_column in column_pairs.items():
    group = data.groupby(column)
    
    if data[column].dtype == 'object':
        # Categorical data - Chi-squared test
        contingency_table = pd.crosstab(data[column], data[kpi_column])
        chi2, p, _, _ = chi2_contingency(contingency_table)
        print(f"Chi-squared test p-value for '{column}' vs. '{kpi_column}': {p}")

        # Analyze p-value to reject/nullify the null hypothesis

    else:
        # Numerical data - T-test
        group_a = group.get_group('Male')  
        group_b = group.get_group('Female') 
        t_stat, p = ttest_ind(group_a[kpi_column], group_b[kpi_column])
        print(f"T-test p-value for '{column}' vs. '{kpi_column}': {p}")
```

```python
column_pairs = [
    ('IsVATRegistered', 'AccountType'),
    ('IsVATRegistered', 'MaritalStatus'),
    ('IsVATRegistered', 'Gender'),
    ('AccountType', 'MaritalStatus'),
    ('AccountType', 'Gender'),
    ('MaritalStatus', 'Gender'),
    ('Gender', 'AlarmImmobiliser'),
    ('Gender', 'TrackingDevice'),
    ('Gender', 'NewVehicle'),
    ('Gender', 'WrittenOff'),
    ('Gender', 'Rebuilt'),
    ('Gender', 'Converted'),
    ('Gender', 'TermFrequency')
]
```

```python

# Convert boolean columns to numerical type
boolean_cols = [    'IsVATRegistered',
    'AccountType',
    'MaritalStatus',
    'Gender',
    'AlarmImmobiliser',
    'TrackingDevice',
    'NewVehicle',
    'WrittenOff',
    'Rebuilt',
    'Converted',
    'TermFrequency']

data[boolean_cols] = data[boolean_cols].astype(str)

# Statistical Testing for all column pairs
for pair in column_pairs:
    col1, col2 = pair
    contingency_table = pd.crosstab(data[col1], data[col2])

    if all(data[col].dtype == 'object' for col in pair):
        # Categorical data - Chi-squared test
        chi2, p, _, _ = chi2_contingency(contingency_table)
        print(f"Chi-squared test p-value for '{col1}' vs. '{col2}': {p}")

        # Analyze p-value to reject/nullify the null hypothesis

    else:
        # Numerical data - T-test
        t_stat, p = ttest_ind(data[col1], data[col2])
        print(f"T-test p-value for '{col1}' vs. '{col2}': {p}")
```
